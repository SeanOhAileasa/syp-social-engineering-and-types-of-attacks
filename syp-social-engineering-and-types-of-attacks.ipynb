### Phishing and its Variants
***

The most common and obvious attack vectors - P H I S H I N G - "Phishing attacks".

A cyber attack that uses disguised email as a vector and the goal is to trick the recipient into believing that the message is legitimate (click a link or download an attachment).

This can happen in corporate email, personal email, web mail, any form of electronic mail.

Email phishing attacks or hoaxes are one of the most common exploit vectors available to crackers. 

< [Table of Contents](#top) | [References](#references) >
<a id="PhishinganditsVariantsVariantsSpearphishing"></a>
***
###### Variants - Spear phishing
***

"Spear phishing" is when you target certain employees, for example, someone in the finance department or the accounting department, maybe a database administrator or a server administrator or a low level employee, there might be more likely to respond to a social engineering attack like a receptionist or a secretary.

< [Table of Contents](#top) | [References](#references) >
<a id="PhishinganditsVariantsVariantsWhaling"></a>
***
###### Variants - Whaling
***

Taking spear phishing to the next level is "Whaling".

That's where you're going to target high level employees, for example, someone from the C suite or the C team - the CEO, the CIO, the chief financial officer, CFO or senior management. 

It could be somebody on the board of directors. 

< [Table of Contents](#top) | [References](#references) >
<a id="PhishinganditsVariantsVariantsVishing"></a>
***
###### Variants - Vishing
***

Then we have "Vishing" with a V.

"Vishing" uses the same process as phising, however, it targets cell phones, telephones and voiceover IP systems. 

It uses that as the vector instead of email.

Often we call the attacker a visher - V I S H E R. 

The Visher may call spoofing the collection agency or claiming to be a lawyer trying to get personal information or personal health information or intellectual property. 

< [Table of Contents](#top) | [References](#references) >
<a id="PhishinganditsVariantsVariantsSmishing"></a>
***
###### Variants - Smishiing
***

"Smishing" is all of the above, except we're using SMS texting as the vector, instead of email or a telephone.

<a id="PhishinganditsVariantsIndicatorsPhishingAttacks"></a>
***
###### Indicators - Phishing Attacks
***

There are some key indicators to look for in email and web mail to identify phishing attacks. 

Might see "Vague salutations", for example, ``dear valued customer`` or ``dear employee``.

Suspicious-looking domain names, display names, for example, the company name is actually farther down into the url path or the domain is a common misspelling. 

Need to look carefully at all of the text in the url path, often companies will not register the wildcard domain, for example, asterisk, dot whatever dot com and phishers can use modified URLs to trick people into clicking on the links or typing those into the address bar. 

When you move or hover your mouse over hypertext links, If you look down at the bottom, you might see indicators of wrong information or actually just a random IP version for address.

Historically, phishing attacks would have awkward grammar and misspelled words, for example, somebody wrote the phishing email and English wasn't their first language.

However, with tools like [Grammarly](https://www.grammarly.com/) this is becoming a lot less common. 

Sometimes the subject line of an email phishing attack has urgent or intimidating phrases like you must act immediately or attention or urgent.

There's often a lack of legitimate contact information in a phishing email as well, along with spoofed headers and logos and corporate graphics.

<a id="PhishinganditsVariantsIndicatorsPhishingAttacksCommonPhishingAttacksBusinessEmailCompromiseBEC"></a>
***
###### Common Phishing Attacks - Business Email Compromise - BEC
***

One of the most prevalent types of phishing attacks is will be called "BEC" "Business E-mail Compromise". 

This is a form of attack that targets companies who outsource, conduct wire transfers and have suppliers abroad. 

They often target corporate email accounts of high level employees, either spoofed or compromised through tools known as keyloggers or other phishing attacks to perform fraudulent transfers.

In 2016, BEC attacks led to an average of $140,000 in losses for companies globally, and they can be ingenious in spoofing real internal e-mail accounts. 

< [Table of Contents](#top) | [References](#references) >
<a id="PhishinganditsVariantsIndicatorsPhishingAttacksCommonPhishingAttacksBusinessEmailCompromiseBECSchemes"></a>
***
###### Common Phishing Attacks - Business Email Compromise (BEC) - Schemes
***

"Common BEC Schemes" would be "Phony invoices and transfers". 

Companies with foreign suppliers are often targeted with this tactic. 

Fraud of the CEO or the "C-suite" or "C-Team", where an attacker will spoof a company's CEO or a CFO and send an email to employees in particular departments, for example, finance or accounting requesting a money transfer.

Obviously, "Email or webmail account compromise". 

Companies often do phishing campaigns in their own companies to further train and raise security awareness for their employees so they don't respond to the BEC schemes. 

They may impersonate attorneys or members of a legal team or a law firm - may say they're in charge of critical and confidential information.

"Data theft of personally identifiable information (PII)", personal health information, PHI and intellectual property. 

This information may be reconnaissance or information gathering that's going to be used in a future advanced persistent threat.

<a id="PharmingBlendofthewordsPhishingandFarming"></a>
***
###### Pharming - Blend of the words Phishing and Farming -  Name Resolution or DNS
***

"Pharming" is a blend of the words "phishing" and "farming". 

Describes a type of cybercrime that's like phishing. 

With pharming a Web site's traffic is manipulated or spoofed and confidential information is stolen. 

Attackers may install a virus or a trojan on a target that changes the computer hosts file to direct traffic away from its intended target and toward a fake or hoax web site.

Crackers may also poison a DNS server to redirect multiple users to unintentionally go to the fake site, which in turn can be used to install malware on the victim's computer.

### Spam Attacks
***

"Spam" is a slang term, it's a term we use for unsolicited commercial e-mail or junk e-mail. 

E-mail spam is one of the most common denial of service activities and is typically followed into several categories. 

"Advertisements" for one, that's the most common type of spam, offering products and services like weight loss pills and software. 

In many scenarios, it may be a scam, but it may be real.

There's "Chain letters", for example, something bad will happen to you if you don't pass along this chain letter to ten people. 

"Spoofed e-mail related to phishing campaigns". 

Some spam are "Hoaxes or money scams". 

Some spammers might try to grab your attention and then direct you to a malicious web site where there's drive by malware. 

There's "money scams". Those are common, for example, the Nigerian prince scam or money for hungry children? 

Spam could be a "Malware warning, for example, a malware infection on your device like ransomware or a virus and "Unwanted pornography related e-mail".

It could be a search engine crawler or it could be a tool trying to extract information or exfiltrate data from your website.

<a id="SpamoverInstantMessagingSPIM"></a>
***
###### Spam over Instant Messaging - SPIM
***

"SPIM" stands for "Spam over instant messaging", and it refers to unsolicited instant messages like Yahoo! messenger or MSN Messenger or Cisco Jabber.

SPIM disrupts chatting and can also contained viruses or spyware.

Blocking any messages from sources not on your contact list can help prevent SPIM. 

Most anti-virus programs include SPAM and SPIM protection features.

<a id="typosquatting"></a>
***
###### Typosquatting - URL Hijacking - Sting Sites - Fake URL
***

"Typosquatting" involves sitting on sites under someone else's brand or copyright and targeting Internet users who erroneously type a Web site address into their browser address bar.

Other names for "Typosquatting" are URL hijacking, sting sites, or fake URL. 

Some examples would be ``facebool`` instead of ``Facebook``, ``gooogle`` with an extra ``O`` or ``amason`` spelled with an ``S`` instead of a ``Z``.

Most organizations are large companies will also register multiple variants and misspellings of their corporate name - this can help combat typosquatting.
